```python
def list_all_subdir(dirname, folder):
    subdir_list = []
    folder_path = os.path.join(dirname, folder)
    # Credit to https://qr.ae/TWXs8S
    for root, directories, filenames in os.walk(folder_path):
        for directory in directories:
            subdir_list.append(os.path.join(root, directory))
        # for filename in filenames:
            # print (os.path.join(root,filename))
            # subdir_list.append(os.path.join(root,filename))
    subdir_list.append(folder_path)
    if not subdir_list:
        subdir_list.append(folder_path)

    return subdir_list


def get_list_of_input_files(dirname, subdir_list, low, high, full=False):
    """
    Return list of .pdf and .PDF files in /data/ folder
    """
    files_grabbed = []
    for dir in subdir_list:
        types = (dir + '/*.txt', dir + '/*.TXT')

        for files in types:
            files_grabbed.extend(glob.glob(os.path.join(dirname, files)))

    # Exception to handle empty input folder
    if files_grabbed == []:
        print('No input files detected!')
        exit()

    if full:
        return files_grabbed

    return files_grabbed[low:high]
```

```python
def get_num_quotes(sent):
#     return sent.count('"')
    num_quotes = sent.count('“') + sent.count('”')
#     num_quotes = sent.count('“') + sent.count('”')
    return num_quotes
    
def add_tag_to_dict(sent_index, sent, somedict, tag):
    global total_quotes_count
    somedict[sent_index] = (tag, total_quotes_count)
    return somedict
    
def is_quote_start_of_sent(sent):
    # must have capitalized alphabet after quote
#     return bool(re.match('^"[A-Z]', sent))
#     return bool(re.match('^&ldquo;[A-Z]', sent))

    return bool(re.match('“', sent))


def is_quote_end_of_sent(sent):
#     return bool(re.match('.*"$', sent))
#     return bool(re.match('.*&rdquo;$', sent))

    return bool(re.match('”', sent))
    

def is_sent_dialogue_s(sent_index):
    global sent_dia_dict
    return sent_dia_dict[sent_index]=='dialogue-S'

def is_sent_dialogue_p(sent_index):
    global sent_dia_dict
    return sent_dia_dict[sent_index]=='dialogue-P'

def tag_dialogue(sent_index, sent):
    global total_quotes_count, sent_dia_dict, sent_list
    
#     sent = sent.replace("'", '"')
#     sent = sent.replace('“', '"').replace('”', '"')
    
    num_quotes = get_num_quotes(sent)
    
    if num_quotes == 0:
        if (total_quotes_count%2) == 0:
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'not')
        else:
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-P')
    
    elif num_quotes == 1:
        if is_quote_start_of_sent(sent):
            total_quotes_count += 1
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-S')
        elif is_quote_end_of_sent(sent):
            total_quotes_count += 1
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-E')
        else:
            if (total_quotes_count%2) == 0:
                total_quotes_count += 1
                add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-S')
            else: 
                total_quotes_count += 1
                add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-E')
                
    elif num_quotes == 2:
        if (total_quotes_count%2) == 0:
            total_quotes_count += 2
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-F')
        else:
            total_quotes_count += 2
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-E-S')
        
    elif num_quotes > 2:
        if (total_quotes_count%2) == 0:
            total_quotes_count += num_quotes
            if (num_quotes%2) == 0:
                add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-F')
            else:
                add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-F-S')
        else:
            total_quotes_count += num_quotes
            add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-S')

    return sent_dia_dict
    
```

```python
def tag_second_round(sent_row):
    global new_sent_list
    index = sent_row['sent_index']
    if sent_dia_dict[index][0] == 'not':
        return -1
    elif sent_dia_dict[index][0] == 'dialogue-S':
        while sent_dia_dict[index][0] != 'dialogue-E':
            index += 1
        return index
    elif sent_dia_dict[index][0] == 'dialogue-P':
        while sent_dia_dict[index][0] != 'dialogue-E':
            index += 1
        return index
    elif sent_dia_dict[index][0] == 'dialogue-E':
        return index
    elif sent_dia_dict[index][0] == 'dialogue-E-S':
        if ((sent_dia_dict[index-1][0] == 'dialogue-S') | (sent_dia_dict[index-1][0] == 'dialogue-P')):
            return index
        elif ((sent_dia_dict[index-1][0] == 'dialogue-E') | 
              (sent_dia_dict[index-1][0] == 'not') |
              (sent_dia_dict[index-1][0] == 'dialogue-F')):
            while sent_dia_dict[index][0] != 'dialogue-E':
                index += 1
            return index
        else:
            return -1
    elif sent_dia_dict[index][0] == 'dialogue-F':
        return index
    elif sent_dia_dict[index][0] == 'dialogue-F-S':
        while sent_dia_dict[index][0] != 'dialogue-E':
            index += 1
        return index
              
```

```python
df['dialogue_end_index'] = df.apply(lambda x: tag_second_round(x), axis=1)
```
