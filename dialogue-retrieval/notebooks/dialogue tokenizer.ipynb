```python
def list_all_subdir(dirname, folder):
    subdir_list = []
    folder_path = os.path.join(dirname, folder)
    # Credit to https://qr.ae/TWXs8S
    for root, directories, filenames in os.walk(folder_path):
        for directory in directories:
            subdir_list.append(os.path.join(root, directory))
        # for filename in filenames:
            # print (os.path.join(root,filename))
            # subdir_list.append(os.path.join(root,filename))
    subdir_list.append(folder_path)
    if not subdir_list:
        subdir_list.append(folder_path)

    return subdir_list


def get_list_of_input_files(dirname, subdir_list, low, high, full=False):
    """
    Return list of .pdf and .PDF files in /data/ folder
    """
    files_grabbed = []
    for dir in subdir_list:
        types = (dir + '/*.txt', dir + '/*.TXT')

        for files in types:
            files_grabbed.extend(glob.glob(os.path.join(dirname, files)))

    # Exception to handle empty input folder
    if files_grabbed == []:
        print('No input files detected!')
        exit()

    if full:
        return files_grabbed

    return files_grabbed[low:high]
```

```python
# def get_num_quotes(sent):
#     num_quotes = sent.count('&ldquo;') + sent.count('&rdquo;')
#     return num_quotes

def add_tag_to_dict(sent_index, sent, somedict, tag):
    global total_quotes_count
    somedict[sent_index] = (tag, total_quotes_count)
    return somedict

def is_open_quote_in_sent(sent):
    return '“' in sent

def is_close_quote_in_sent(sent):
    return '”' in sent

def is_complete_quote_in_sent(sent):
    return (is_open_quote_in_sent(sent) & is_close_quote_in_sent(sent))

def tag_dialogue(sent_index, sent):
    global sent_dia_dict
    
    if is_open_quote_in_sent(sent):
        sent_dia_dict = add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-S')
    if is_close_quote_in_sent(sent):
        sent_dia_dict = add_tag_to_dict(sent_index, sent, sent_dia_dict, 'dialogue-F')
    else:
        sent_dia_dict = add_tag_to_dict(sent_index, sent, sent_dia_dict, 'not')

    return sent_dia_dict
    
```

```python
# total_quotes_count = 0
sent_dia_dict = dict()
for sent_index, sent in enumerate(paragraphs):
    sent_dia_dict = tag_dialogue(sent_index, sent)
```

```python
df = pd.DataFrame(list(sent_dia_dict.items()), columns=['sent_index', 'tag'])
```

```python
# df['num_quotes_up_to_sent'] = df['tag'].apply(lambda x: x[1])
df['tag'] = df['tag'].apply(lambda x: x[0])
```

```python
df['sent'] = [x.strip().replace('\r\n', '') for x in paragraphs]
```
